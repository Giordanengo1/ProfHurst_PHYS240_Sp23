```python
import numpy as np
import matplotlib.pyplot as plt
```

Problem 1: Plotting Boltzmann Distribution

```python
def P_v(v):
    result = 4*np.pi*(1/(2*np.pi))**(3^2)*v**2*np.exp(-0.5*(v**2))
    return result #result is not normalized as all consants are set to 1

Prob = np.zeros(100)
xplot = np.linspace(0, 5, num=100) # a = 0 , b=5

#for i in r
Prob = P_v(xplot)

plt.plot(xplot, Prob) #Boltzmann distribution

print('The maximum value is:', max(Prob))
```

Code for Problem 2. P_Max value is chosen to be 1.48

```python

rng = np.random.default_rng(12345)  #random number generator
samples = 1000

r1 = rng.random(size=samples)   #two R-values
r2 = rng.random(size=samples)

#print(r1)
#print(r2)

x_try = 5*r1     #formula reduces down with a=0
y_try = 1.48*r2  # P_max = 1.48

AcceptedX = np.zeros(samples)   #Arrays for binning the in/valid coordinates
RejectedX = np.zeros(samples)   
AcceptedY = np.zeros(samples)
RejectedY = np.zeros(samples)

NumAcc = 0 #Variables double as counter and index
NumRej = 0

for i in range (samples-1):
    if P_v(x_try[i]) > y_try[i]:    #Accepted result
        AcceptedX[NumAcc] = x_try[i]
        AcceptedY[NumAcc] = y_try[i]
        NumAcc = NumAcc + 1
    else:                           #Rejected result
        RejectedX[NumRej] = x_try[i]
        RejectedY[NumRej] = y_try[i]
        NumRej = NumRej + 1

AccXPlot = np.zeros(NumAcc+1) #Declare Arrays for plotting, resized for number of respective cases
AccYPlot = np.zeros(NumAcc+1)
RejXPlot = np.zeros(NumRej+1)
RejYPlot = np.zeros(NumRej+1)        
        
AccXPlot = AcceptedX[:NumAcc] #Slice data from binning arrays into plotting arrays
AccYPlot = AcceptedY[:NumAcc]
RejXPlot = RejectedX[:NumRej]
RejYPlot = RejectedY[:NumRej]
```

Problem 3 is below. A graph of our Boltzmann Distribution and all accepted (green) and rejected (red) random points

```python
plt.figure()
plt.plot(RejXPlot,RejYPlot, 'r+')
plt.plot(AccXPlot,AccYPlot, 'g+')
plt.plot(xplot, Prob)
```
